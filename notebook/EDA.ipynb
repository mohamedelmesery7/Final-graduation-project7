```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\data\last_data.csv")
```

```python
df.head()
```

```python
df['Protocol'].value_counts(sort=True).head(10)
```

```python
plot = df['Protocol'].value_counts(sort=True).head(10)
ax = sns.barplot(x = plot.index , y= plot.values)

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=9)

    
    
plt.title('Top 10 protocols')
plt.ylabel('Count')
plt.show()
```
